```python
import numpy as np
from scipy.stats import skew, kurtosis
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

# 1. Generate some sample data

```python
np.random.seed(42)
data=np.random.normal(loc=10,scale=2,size=100)          #normally distributed data
data_skew=np.random.exponential(scale=2,size=100)     #skewed data
data_binomial=np.concatenate([np.random.normal(loc=5,scale=1,size=50),
                                      np.random.normal(loc=15,scale=1,size=50)])   #binomial
```

```python
print("--- Original Data (Normal) ---")
print(f"Mean: {np.mean(data):.2f}")
print(f"Standard Deviation: {np.std(data):.2f}")
```

# 2. Skewness

```python
print("\n--- Skewness ---")
print(f"Skewness of normal data: {skew(data):.2f} (expected close to 0)")
print(f"Skewness of skewed data (exponential): {skew(data_skew):.2f} (expected positive)")
```

# 3. Kurtosis

```python
print("\n--- Kurtosis ---")
# Fisher's kurtosis (excess kurtosis) is usually reported, where normal distribution has 0.
print(f"Kurtosis of normal data: {kurtosis(data):.2f} (expected close to 0 for Fisher's kurtosis)")
print(f"Kurtosis of bimodal data: {kurtosis(data_binomial):.2f} (expected negative for platykurtic)")
```

# 4. Normalization (Min-Max Scaling)

```python
print("\n--- Normalization (Min-Max Scaling) ---")
scaler_minmax = MinMaxScaler()
data_normalized = scaler_minmax.fit_transform(data.reshape(-1, 1)).flatten()
print(f"Original Min: {np.min(data):.2f}, Max: {np.max(data):.2f}")
print(f"Normalized Min: {np.min(data_normalized):.2f}, Max: {np.max(data_normalized):.2f}")
```

# 5. Standardization (Z-score Scaling)

```python
print("\n--- Standardization (Z-score Scaling) ---")
scaler_standard = StandardScaler()
data_standardized = scaler_standard.fit_transform(data.reshape(-1, 1)).flatten()
print(f"Original Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
print(f"Standardized Mean: {np.mean(data_standardized):.2f} (expected close to 0)")
print(f"Standardized Std Dev: {np.std(data_standardized):.2f} (expected close to 1)")
```

# 6. Correlation

```python
print("\n--- Correlation ---")
# Create two correlated variables
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100) # y is positively correlated with x
z = -3 * x + np.random.normal(0, 0.1, 100) # z is negatively correlated with x
```

```python
correlation_xy = np.corrcoef(x, y)[0, 1]
correlation_xz = np.corrcoef(x, z)[0, 1]
```

```python
print(f"Correlation between x and y (positive): {correlation_xy:.2f}")
print(f"Correlation between x and z (negative): {correlation_xz:.2f}")
```

```python
# A more comprehensive correlation matrix
print("\n--- Correlation Matrix ---")
combined_data = np.vstack([x, y, z])
correlation_matrix = np.corrcoef(combined_data)
print("Correlation Matrix:\n", np.round(correlation_matrix, 2))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
# 1. Set the visual style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 10))

# --- GENERATING DATA ---

# A. Normal Distribution (Heights)
# Mean = 170cm, Std Dev = 10, Size = 1000
normal_data = np.random.normal(170, 10, 1000)

# B. Right-Skewed Distribution (Income)
# Using a Gamma distribution to create a long tail to the right
skewed_data = np.random.gamma(2, 2, 1000) * 5000

# C. Bimodal Distribution (Restaurant Traffic)
# Two peaks: Lunch rush (12pm) and Dinner rush (6pm)
lunch = np.random.normal(12, 1, 500)
dinner = np.random.normal(18, 1, 500)
bimodal_data = np.concatenate([lunch, dinner])
```

```python
# --- PLOTTING ---

# Plot 1: Normal Distribution
plt.subplot(2, 2, 1)
sns.histplot(normal_data, kde=True, color="skyblue")
plt.title(f"Normal Distribution (Symmetrical)\nSkew: {skew(normal_data):.2f} | Kurtosis: {kurtosis(normal_data):.2f}")

# Plot 2: Right-Skewed Distribution
plt.subplot(2, 2, 2)
sns.histplot(skewed_data, kde=True, color="salmon")
plt.title(f"Right-Skewed Distribution (Income)\nSkew: {skew(skewed_data):.2f} | Kurtosis: {kurtosis(skewed_data):.2f}")

# Plot 3: Bimodal Distribution
plt.subplot(2, 2, 3)
sns.histplot(bimodal_data, kde=True, color="teal")
plt.title(f"Bimodal Distribution (Two Peaks)\nTwo distinct 'averages' present")

# Plot 4: Comparison Boxplot
plt.subplot(2, 2, 4)
sns.boxplot(data=[normal_data/10, skewed_data/1000], palette="Set2")
plt.xticks([0, 1], ['Normal (Scaled)', 'Skewed (Scaled)'])
plt.title("Boxplot: Notice the Outliers in Skewed Data")

plt.tight_layout()
plt.show()
```

```python

#NOrmalization
import numpy as np
import pandas as pd

data = [10, 20, 30, 40, 50]

df = pd.DataFrame(data, columns=['Values'])

# Min-Max Normalization
df['Normalized'] = (df['Values'] - df['Values'].min()) / (df['Values'].max() - df['Values'].min())

print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df['Normalized'] = scaler.fit_transform(df[['Values']])

print(df)
```

```python
#standardization


import pandas as pd

data = [10, 20, 30, 40, 50]

df = pd.DataFrame(data, columns=['Values'])

# Standardization
df['Standardized'] = (df['Values'] - df['Values'].mean()) / df['Values'].std()

print(df)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['Standardized'] = scaler.fit_transform(df[['Values']])

print(df)
```

```python
#add new column in existing data
mapping = {
    'setosa': 0,
    'versicolor': 1,
    'virginica': 2
}

df['numeric_specis'] = df['species'].map(mapping)
df.head()
```

```python
df['modified_sepal_len'] = df['sepal_length']*10
df.head()
```

```python
df.drop("modified_sepal_len",axis=1,inplace=True)
df.head()
```

```python
df.sort_values(by="sepal_length",ascending=False)  #sort the data by column values
```
